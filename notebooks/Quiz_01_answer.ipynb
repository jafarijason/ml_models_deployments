Question. The marketing manager at X company, an lawn mower company, belives that monthly sales across all outlets(stores, online, etc.). are influenced by three key varibles:
  1. outdor tempreture (in F)
  2. advertising expenditures (in $1000s)
  3. promotional discount (in %)
The accompanying data frame shows monthly sales data over the past two years


```python
import pandas as pd
import matplotlib.pyplot as plt

mowersDf = pd.read_excel('https://www.dropbox.com/scl/fi/y2rktyoqb8rrshrnlpvw1/Mowers.xlsx?rlkey=e5bi1d8sx5hml4ylfkjv7cryh&dl=1')
mowersDf.head()
```

}  $$
and test for the joint and indivisual significant of predictor variables at the 5% level.

```python
# https://www.statsmodels.org/stable/index.html
import statsmodels.api as sm
# Your answer
```

```python
model1 = sm.OLS(
  mowersDf["Sales"],
  sm.add_constant(mowersDf[["Temperature", "Advertising", "Discount"]])
)
model1Fit = model1.fit()
print(model1Fit.summary())
```
